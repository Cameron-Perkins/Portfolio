```python
%%writefile CudaSorting.cu
#include <cuda_runtime_api.h>
#include <memory.h>
#include <cstdlib>
#include <ctime>
#include <stdio.h>
#include <cmath>
#include <iostream>
#include <random>

void initArray(float* A, int length){
    for(int i=0; i<length; i++){
        A[i] = 10 - i;
    }
}

void printArray(float* A, int length){
    for(int i=0; i<length; i++){
        std::cout << A[i] << ' ';
    }
}

__global__ void check_if_array_is_sorted(float *A, float *B,int N){

    // accessing the block, we also need the dimentions of the block, then we get the specific thread
    int i = blockIdx.x * blockDim.x + thread.x;
    // First we compare each left element to its right
    if(i < N){
        if (A[i] < A[i+1]){
            B[i] = 0;
        }
        else{
            B[i] = 1;
        }
    }

    int ans;
    if(i<N){
        B[0] += B[i];
    }
}

bool cpu_check_if_array_is_sorted(float *A, float *B, int N){

    for(int i = 0; i<N-1; i++){
        if(A[i] > A[i+1]){
            return false;
        }
    }
    return true;
}

void run_cpu() {

    int vectorLength = 10;
    // I need to allocate CPU memory
    float* A = nullptr;
    float* B = nullptr;
    
    // Pointers for device
    float* devA = nullptr;
    float* devB = nullptr;
    
    // Allocate host memory
    cudaMallocHost(&A, vectorLength);
    cudaMallocHost(&B, vectorLength);

    // Then initialize the vectors
    initArray(A, vectorLength);
    initArray(B, vectorLength);

    printArray(A, vectorLength);

    cudaMalloc(&devA, vectorLength*sizeof(int));
    cudaMalloc(&devB, vectorLength*sizeof(int));
    printArray(A, vectorLength);
    
    // Then copy the data
    int threads = 256;
    int blocks = (vectorLength + threads - 1) / threads;
    check_if_array_is_sorted<<<blocks, threads>>>(devA, devB, vectorLength);

    cudaDeviceSynchronize();

    cudaMemcpy(B, devB, vectorLength*sizeof(float), cudaMemcpyDefault);

    if (B[0] == 0){
        std::cout << "sorted";
    }
    else{
        std::cout << "not sorted";
    }
    

    cudaFree(devA);
    cudaFree(devB);
    cudaFreeHost(A);
    cudaFreeHost(B);
}

int main(){
    run_cpu();
    return 0;
}
```
